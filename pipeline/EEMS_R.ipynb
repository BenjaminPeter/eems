The first thing we want to do is to load the R plotting scripts. I then define some simple wrappers for Desis plotting functions. They should be expanded/made more customizable

```python
mcmcpath = "/data/eems-project/eems/output/eems_opt"
datapath = "/data/eems-project/eems/python-pipeline/test"

mcmcpath = os.path.sep.join( (options['tmp_dir'], options['output_files'] ) )
datapath = os.path.sep.join( (options['tmp_dir'], options['input_files'] ))

def load_r_plotting_functions():
    %R source("/data/eems-project/eems/rscripts/src/myeems.plots.R")
    %Rpush mcmcpath datapath
    %R dimns <- read.dimns(datapath)
    

def generic_plotter( plotter, target="screen", width=4, height=4, close=False, **kwargs):
    %Rpush target width height close
    kwdict = ListVector(kwargs)
    %Rpush kwdict plotter
    %R generic_plotter( plotter, target, kwdict)
    
def plot_fitted_vs_observed_distance( target=mcmcpath + "_fitted_observed_distance.pdf" ):
    %Rpush target
    %R generic_plotter( dist.scatterplot, list( mcmcpath ), target )
    
def plot_max_migration_surface(target=mcmcpath + "_max_migration.pdf"):
    %Rpush target
    %R mlegend <- generic_plotter( mcmc.mrates, list(mcmcpath,dimns), target=target, close=F)
    %R generic_plotter( mcmc.mrates.legend, plotter_args=list(datapath,mcmcpath,dimns,mmrks,mlegend) )
    

def plot_max_population_size_surface(target=mcmcpath + "_max_population_size.pdf"):
    %Rpush target
    %R qlegend <- generic_plotter( mcmc.qrates, list(mcmcpath,dimns), target=target, close=F)
    %R generic_plotter( mcmc.qrates.legend, plotter_args=list(datapath,mcmcpath,dimns,qmrks,mlegend) )
    
def plot_trace(target=mcmcpath + "_max_population_size.pdf"):
    %Rpush target
    %R pilogl <- read.table(paste(mcmcpath,'.mcmcpilogl',sep=''))
    %R post <- pilogl[,1] + pilogl[,2]
    %R generic_plotter( plot, list("x"=post, "type"="l", ylab=""), target=target )

load_r_plotting_functions()
```

### Migration surface

```python
plot_max_migration_surface()
plot_max_migration_surface(target="screen")
```

### Effective population size surface

```python
plot_max_population_size_surface()
plot_max_population_size_surface(target="screen")
```

### MCMC trace plot

```python
plot_trace()
plot_trace(target="screen")
```

### Fitted vs observed distances

```python
plot_fitted_vs_observed_distance()
plot_fitted_vs_observed_distance(target="screen")
```
